```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
import pandas as pd
import chardet

# Path to the uploaded CSV file
csv_file_path = 'Articles.csv'

# Detect the encoding of the file
with open(csv_file_path, 'rb') as f:
    result = chardet.detect(f.read())

file_encoding = result['encoding']
print(f"Detected file encoding: {file_encoding}")

# Read the CSV file with the detected encoding
df = pd.read_csv(csv_file_path, sep=',', quotechar='"', encoding=file_encoding, engine='python')

# Display the first few rows of the DataFrame to verify contents
print(df.head())

# Optional: Save the cleaned DataFrame to a new CSV file
output_file_path = 'Cleaned_Articles.csv'
df.to_csv(output_file_path, index=False)

print(f"Cleaned CSV file saved to: {output_file_path}")
```

```python


df.columns = ['content', 'publish_date', 'headline', 'category']
df['publish_date'] = pd.to_datetime(df['publish_date']).dt.strftime('%Y-%m-%d')

df
```

```python
engine = create_engine('postgresql+psycopg2://postgres:gajah@localhost:5432/mediom')

# Insert the DataFrame into the PostgreSQL table
table_name = 'article'
df.to_sql(table_name, engine, if_exists='append', index=False, method='multi')

print(f"Data inserted into the '{table_name}' table successfully.")
```
